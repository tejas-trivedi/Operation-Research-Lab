```python
from scipy.optimize import linprog
import time
c = [-3,-2,-3]
A = [[2,1,1],[1,3,1],[3,4,2]]
b = [2,6,8]
x1_bnds = (0, None)
x2_bnds = (0, None)
x3_bnds = (0, None)
begin = time.time()
res = linprog(c, A_eq=A, b_eq=b, bounds=(x1_bnds, x2_bnds, x3_bnds), method='simplex', options={"disp":True})
print("Time taken: %s seconds"%(time.time()-begin))
print(res)
```
